#### Dictionaries

```python
# Personalized greetings using a dictionary
greetings = {
    "world": "Hello",
    "Alice": "Hi",
    "Bob": "Hey"
}

for name, greeting in greetings.items():
    print(f"{greeting}, {name}!")
```

- This is a collection of key-value pairs, keys must be unique. 
- The importance of dictionaries include the speed, the code allows looking up items to become an easy task, finding items through a list is made easy. 
- The use of values allows for data to become self-describing, creating clear messaging.

```python
# Using array from the standard library for numbers
from array import array

counts = array("i", [1, 2, 3])  # "i" means signed integer
for c in counts:
    print(f"Hello, world! x{c}")
```

#### Conditionals (if, elif, else)

```python
# Hello with different messages based on time of day
time_of_day = "morning"  # try "afternoon" or "evening"

if time_of_day == "morning":
    print("Good morning, world!")
elif time_of_day == "afternoon":
    print("Good afternoon, world!")
else:
    print("Good evening, world!")
```

#### Classes, objects, and methods

```python
# A simple Greeter class
class Greeter:
    def __init__(self, greeting):
        self.greeting = greeting  # instance attribute

    def greet(self, target):
        return f"{self.greeting}, {target}!"

greeter = Greeter("Hello")   # create an object
print(greeter.greet("world"))
```

### *example 1*

```python
import string
from collections import Counter

# Poem text 
poem = """
Boil water
Place teabag in mug
Wait 2 Minutes
Remove teabag
Add milk if desired
Output tea

"""

# Step 1: Normalize and split text into words
words = [w.lower().strip(string.punctuation) for w in poem.split()]

# Step 2: Count word frequencies
counts = Counter(words)

# Step 3: Find the most frequent word
most_common = counts.most_common(1)[0]

# Step 4: Output result
print("Most frequent word:", most_common[0], "appears", most_common[1], "times")
```

# Understanding a Program: Parsing and Reading

```python
import webbrowser
import requests

print("Shall we hunt down an old website?")

site = input("Type a website URL: ")

era = input("Type year, month, and date, e.g., 20150613: ")

url = f"https://archive.org/wayback/available?url={site}&timestamp={era}"


response = requests.get(url)
data = response.json()

try:
    old_site = data["archived_snapshots"]["closest"]["url"]
    print("Found this copy:", old_site)
    print("It should appear in your browser.")
    webbrowser.open(old_site)
except:
    print("Sorry, could not find the site.")
```
